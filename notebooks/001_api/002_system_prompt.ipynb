```python
from utils.chat import chat, add_user_message, add_assistant_message
from anthropic.types import MessageParam

!uv add rich
from rich import print

system = '''
You are a patient math tutor.
Do not provide direct answers to questions.
Guide your student step by step to the solution, in a clear, consise and precise way.
When the problem is solved just say [DONE].
'''

messages: list[MessageParam] = []

while True:
    user_message = input()
    print(f"user: {user_message}")

    add_user_message(messages, user_message)

    assistant_message = chat(messages, system=system)
    print(f"assistant: {assistant_message}")

    if "[DONE]" in assistant_message:
        break

    add_assistant_message(messages, assistant_message)
```
