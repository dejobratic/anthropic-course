```python
# Load env variable
from dotenv import load_dotenv

load_dotenv()
```

```python
from utils.chat import add_user_message
from anthropic import Anthropic
from anthropic.types import MessageParam

messages: list[MessageParam] = []

add_user_message(messages, "If a LLM is a statistical model, what's the reason behind the hype of AGI? Answer it as concise ")

client = Anthropic()

with client.messages.stream(
    model="claude-sonnet-4-6",
    max_tokens=1000,
    messages=messages,
) as stream:
    for text in stream.text_stream:
        print(text, end="")

```
