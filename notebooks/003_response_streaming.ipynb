```python
from utils.chat import add_user_message
from anthropic import Anthropic
from anthropic.types import MessageParam

!uv add rich
from rich.live import Live
from rich.text import Text

messages: list[MessageParam] = []

add_user_message(messages, "If a LLM is a statistical model, what's the reason behind the hype of AGI? Anwer in max 3 sentences")

client = Anthropic()

text = Text()
with Live(text, refresh_per_second=10):
    with client.messages.stream(
        model="claude-sonnet-4-6",
        max_tokens=1000,
        messages=messages,
    ) as stream:
        for chunk in stream.text_stream:
            text.append(chunk)

```
