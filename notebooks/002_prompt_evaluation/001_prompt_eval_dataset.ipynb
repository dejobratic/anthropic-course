````python
# Load env variables
from dotenv import load_dotenv

load_dotenv()
````

````python
from utils.chat import chat, add_assistant_message, add_user_message
from anthropic.types import MessageParam
import json


def generate_dataset() -> str:
    prompt = """
Generate a evaluation dataset for a prompt evaluation. The dataset will be used to evaluate prompts
that generate Python, JSON, or Regex specifically for AWS-related tasks. Generate an array of JSON objects,
each representing task that requires Python, JSON, or a Regex to complete, together with the solution criteria.

Example output:
```json
[
    {
        "task": "Description of task",
        "format": "json" or "python" or "regex",
        "solution_criteria": "concise and precise description of solution criteria"
    },
    ...additional
]
```

* Focus on tasks that can be solved by writing a single Python function, a single JSON object, or a regular expression.
* Focus on tasks that do not require writing much code
* Focus on meaningful solution criteria, like performance, memory etc.

Please generate 3 objects.
"""
    messages: list[MessageParam] = []

    add_user_message(messages, prompt)
    add_assistant_message(messages, "```json")
    text = chat(messages, model="claude-haiku-4-5", stop_sequences=["```"])
    return json.loads(text)
````

````python
answer = generate_dataset()

with open("002_prompt_evaluation/dataset.json", "w") as f:
    json.dump(answer, f, indent=2)

answer
````
